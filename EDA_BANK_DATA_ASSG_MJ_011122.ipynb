```python
#lets seperate it by creating a new dataset

Numerical_dataset = APDT[Numerical_columns]

#but this data contains mixed target values, lets seperate defaulters and non-defaulters
Numerical_defaulters = Numerical_dataset[Numerical_dataset['TARGET']==1]
Numerical_nondefaulters = Numerical_dataset[Numerical_dataset['TARGET']==0]
Numerical_defaulters
Numerical_nondefaulters
```
